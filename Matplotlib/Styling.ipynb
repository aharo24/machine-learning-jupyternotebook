```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label="X vs X")

ax.plot(x,x**2, label="X vs X^2")

ax.legend(loc=(1.01,0.5))
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#9bb5a7', label="greish")

ax.plot(x,x+1, color='red', label="red")

ax.legend()
```
